```python
import requests
import bs4,openpyxl
```

# Creating EXCEL SHEET 

```python
exel = openpyxl.Workbook()
print(exel.sheetnames)
sheet = exel.active
sheet.title = 'Imdb movie top rated'
print(sheet.title)
sheet.append(['Rank','Movie Name','Year','Rating'])
```

# Target URL

```python
url = ('https://www.imdb.com/chart/top/')
```

Getting Request From Site

```python
req = requests.get(url)
req.raise_for_status
```

# Let's Create BeautifulSoup

```python
soup = bs4.BeautifulSoup(req.text,'lxml')
```

```python
movies = soup.find('tbody', class_=('lister-list')).find_all('tr')
```

## Grabbing All The Titles By Using Tags & looping

```python
for movie in movies:
    name = movie.find('td', class_=('titleColumn')).a.text
    rank = movie.find('td', class_=('titleColumn')).get_text(strip=True).split('.')[0]
    year = movie.find('td', class_=('titleColumn')).span.text.strip('()')
    rating = movie.find('td',class_=('ratingColumn imdbRating')).text
    print(rank,name,year,rating)
    sheet.append([rank,name,year,rating])
```

# # Now Creating and Naming Excel File With EXtention

```python
exel.save('Top 250 Movies Rated on IMDB.xlsx')
```
